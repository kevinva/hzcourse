```python
from enum import Enum, unique
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import lightgbm as lgb

RANDOM_SEED = 6666
```

```python
def encode_tag(tag_list):
    tag_dict = {}
    for index, tag in enumerate(tag_list):
        tag_dict[tag] = index + 1
    return tag_dict
```

```python
manage_tags = ["问题按期办结", "督导检查", "河长会议", "下级河长履职", "工作奖惩", "下级管理"]
encode_tag(manage_tags)
```

```python
knowledge_tags = ["河湖风险", "巡河覆盖", "问题办理质量", "履职评价", "问题上报", "巡河轨迹"]
encode_tag(knowledge_tags)
```

```python
def data_preprocess(df, label_col):
    features = ["level", "will_score", "manage_score", "knowledge_score", "habit_score", "skill_score"]
    X_data = df[features].copy()
    
    level_encoder = LabelEncoder()
    
    X_data["level_encoded"] = level_encoder.fit_transform(X_data["level"])
    X_data.drop(["level"], axis = 1, inplace = True)
    
    X_data_list = []
    y_data_list = []
    for i in range(len(X_data)):
        item_info = X_data.iloc[i].to_dict()
        tag_list = df[label_col].iloc[i]
        for tag in tag_list:
            X_data_list.append(item_info.copy())
            y_data_list.append(tag) 
    
    X_data_new = pd.DataFrame(X_data_list)
        
    tag_dict = None
    if label_col == "manage_tag":
        tag_dict = encode_tag(manage_tags)
    elif label_col == "knowledge_tag":
        tag_dict = tag_encoder(knowledge_tags)
        
    tag_encoder = None
    if tag_dict is None:
        print("tag_dict is None")
        return None
    
    tag_encoder = LabelEncoder()
    tag_encoder.classes_ = [key for key in tag_dict.keys()]
    tag_encoder.transform = lambda x: [tag_dict[value] for value in x]
    
    y_data_new = tag_encoder.fit_transform(y_data_list).tolist()
    
    X_train, X_eval, y_train, y_eval = train_test_split(X_data_new, y_data_new, test_size = 0.1, stratify = y_data_new, random_state = RANDOM_SEED)
    X_test, X_eval, y_test, y_eval = train_test_split(X_eval, y_eval, test_size = 0.5, stratify = y_eval, random_state = RANDOM_SEED)
    return X_train, X_eval, X_test, y_train, y_eval, y_test
    
```

```python
X_train, X_eval, X_test, y_train, y_eval, y_test = data_preprocess(data_table, "manage_tag")
```

```python
g_lightGBM_params = {
    'max_depth': 5,   # max_depth也需要调参，但先固定比较好调（num_leaves跟这个有关）
    'num_leaves': 27,
    'min_data_in_leaf': 19,
    'feature_fraction': 0.6,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'lambda_l1': 0.3,
    'lambda_l2': 0.03,
    'pred_threshold': 0.5
}

@unique
class ALGORITHM(Enum):
    LGB = 'LGB'
    CTB = 'CTB'
    XGB = 'XGB'
```

```python
def train(X_train, X_eval, y_train, y_eval, num_class, algorithm = ALGORITHM.LGB):
    print('#################### Start training ####################')
    
    model = None
    start_time = time.time()
    
    if algorithm == ALGORITHM.LGB:
        lgb_train = lgb.Dataset(X_train, y_train)
        lgb_eval = lgb.Dataset(X_eval, y_eval, reference = lgb_train)

#         params = {
#             'task': 'train',
#             'boosting_type': 'gbdt',
#             'objective': 'multiclass',
#             'metric': 'multi_error',
#             "num_class": num_class,
#             'is_training_metric': True,
#             'verbose': 20,
#             'learning_rate': 0.05,
#             'early_stopping_round': 50, 
#             'num_iterations': 500,
#             'random_state': RANDOM_SEED,

#             'max_depth': g_lightGBM_params['max_depth'],
#             'num_leaves': g_lightGBM_params['num_leaves'],
#             'min_data_in_leaf': g_lightGBM_params['min_data_in_leaf'],
#             'feature_fraction': g_lightGBM_params['feature_fraction'],
#             'bagging_fraction': g_lightGBM_params['bagging_fraction'],
#             'bagging_freq': g_lightGBM_params['bagging_freq'],
#             'lambda_l1': g_lightGBM_params['lambda_l1'],
#             'lambda_l2': g_lightGBM_params['lambda_l2'],
#         }
        
        params = {
            'objective': 'multiclass',
            'metric': ['multi_logloss', 'multi_error'],
            'num_class': num_class,
            'n_jobs': 4,
            'num_iterations': 300,
            'learning_rate': 0.02,
            'max_depth': 4,
            'lambda_l2': 0.8,
            'verbose': -1
        }


        eval_result={}
        model = lgb.train(params = params, 
                          train_set = lgb_train, 
                          valid_sets = lgb_eval, 
                          callbacks=[lgb.log_evaluation, lgb.early_stopping(50, first_metric_only = True), lgb.record_evaluation(eval_result)])
        
        end_time = time.time()

        print('#################### Finish training ####################')
        print(f'耗时: {end_time - start_time}.')
    
    return model, eval_result
```

```python
def predict(model, X_test, y_test, algorithm = ALGORITHM.LGB):
    if algorithm == ALGORITHM.LGB:
#         model = lgb.Booster(model_file = model_path)
        pred_threshold = g_lightGBM_params['pred_threshold']
        y_tmp_pred = model.predict(X_test, num_iteration = model.best_iteration)
        print(y_tmp_pred)
        
        pred_tensor = torch.tensor(y_tmp_pred)
        pred_scores = pred_tensor / torch.sum(pred_tensor, dim = 0)
        print(pred_scores)
        
        y_pred = []
        for values in y_tmp_pred:
            pre_index = np.argmax(values) + 1
            y_pred.append(pre_index)
        
        y_true = np.array(y_test).tolist()
        assert len(y_true) == len(y_pred)
        
        confusion_mat = confusion_matrix(y_true, y_pred)
#         accuracy = accuracy_score(y_true, y_pred)
#         recall = recall_score(y_true, y_pred)
#         precision = precision_score(y_true, y_pred)
        report = classification_report(y_true, y_pred)
        
        print(confusion_mat)
        print(report)
#         print(f"accuracy: {accuracy}, recall: {recall}, precision: {precision}")
```

```python
result_dict = {}
last_character = None
last_ability = None
last_tag = None

for i in range(len(df_data2)):
    row_data = df_data2.iloc[i]
    character = row_data[0]
    if character == "面向人群":
        continue
    
    ability = row_data[1]
    tag = row_data[3]
    course = row_data[5]
    
    if ability == "基础\n新上任基层河长（上任6个月内）" or ability == "基础\n新上任区级河长（上任6个月内）":
        ability = "基础"
    elif ability == "管理属性":
        ability = "管理"
    elif ability == "技术属性（知识）":
        ability = "知识"
    elif ability == "业务属性（技能）":
        ability = "技能"
    elif ability == "意愿属性":
        ability = "意愿"
    elif ability == "履职习惯":
        ability = "习惯"
    
    if last_character != character:
        character_dict = {}
    
    if last_ability != ability:
        ability_dict = {}
    
    if last_tag != tag:
        tag_dict = {}
        
    tag_dict[course] = 0.0
    ability_dict[tag] = tag_dict
    character_dict[ability] = ability_dict
    result_dict[character] = character_dict
    
    last_character = character
    last_ability = ability
    last_tag = tag
```
